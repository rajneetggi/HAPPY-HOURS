```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
import numpy as np
import pandas as pd
import difflib
```

```python
# loading the data from the csv file to pandas dataframe(structured form of data i.e. in the form of table)
```

```python
data = pd.read_csv('movies.csv')
```

```python
# reading the first 6 rows of the dataset 
```

```python
data.head()
```

```python
# checking dimmensions of the dataset  
```

```python
data.shape
```

```python
# selecting the relevant features for recommendation to be given to the user 
```

```python
features =  ['genres','keywords','tagline','cast','director']
```

```python
# changing the null values with null string of each feature that is to be selected for prediction with the help of fillna fucntion
```

```python
for f in features:
  data[f] = data[f].fillna('')
```

```python
# combine  all the  selected features for recommendation 
```

```python
comb_features = data['genres']+' '+data['keywords']+' '+data['tagline']+' '+data['cast']+' '+data['director']
print(comb_features)
```

```python
# convert the raw document(Text) into feature vector 
```

```python
vectorizer = TfidfVectorizer()
```

```python
# fit_transform() method with the combined feature
features_imp= vectorizer.fit_transform(comb_features)
print(features_imp)
```

```python
# cosine similarity algorithm is used to determine the distance between the neighbours in recommendation systems
```

```python
similarity = cosine_similarity(features_imp)
print(similarity)
```

```python
print(similarity.shape)
```

```python
movie_name = input(' Enter your favourite movie name : (For Example : Iron Man , Spider Man , Batman,etc.) : '   )
```

```python
title_list = data['title'].tolist()
print(title_list)
```

```python
closest_match = difflib.get_close_matches(movie_name, title_list)
print("These are the similar match to your choice " , closest_match )
```

```python
try:
    close_match = closest_match[0]
    print("Perfect closest match : - " , close_match)
except:
    print("Movie entered is not present in dataset")
```

```python
movie_ind = data[data.title == close_match]['index'].values[0]
print(movie_ind)

    
```

```python
sim_score = list(enumerate(similarity[movie_ind]))
print("Score of the Similarity: - ")
print()
print(sim_score)
```

```python
sort_Mov= sorted(sim_score, key = lambda x:x[1], reverse = True) 
print(sort_Mov)
```

```python
# print the name of similar movies based on the index

print('Movies suggested for you : \n')

i = 1

for movie in sort_Mov:
  index = movie[0]
  titles= data[data.index==index]['title'].values[0]
  if (i<30):
    print(i, '.',titles )
    i+=1
```
